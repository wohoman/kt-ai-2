```python
#이상치 제거하기 'cust_clas_itg_cd'열의 R을 가지고 있는 값 제거하기 
df_data=df[(df['cust_clas_itg_cd']!='R')]
```

```python
#이상치 제거하기

#Q1, Q3 구하기
q1 = df['열 이름'].quantile(0.25)
q3 = df['열 이름'].quantile(0.75)

# 1.5 * IQR(Q3 - Q1)
iqr = 1.5 * (q3 - q1) 

# 이상치 제거하기
df_data=df[(df['열 이름'] < (q3 + iqr)) & (df['열 이름'] > (q1 - iqr))]

df_data.describe()
```

```python
#이상치 변경하기

# 이상치 대체값 설정하기
age_min = q1 - iqr
age_max = q3 + iqr

# 이상치를 변경
df[(df['열 이름'] > age_max)]= age_max
df[(df['열 이름'] < age_min)]= age_min

df.describe()
```

```python
# 모델 호출하기
# 파라미터 (C: 규제강도, max_iter: 계산에 사용할 작업 수)
lr_model = LogisticRegression()

# 모델 학습하기
lr_model.fit(X_train, y_train)
```

```python
# ② Train / Test 데이터 Set 나누기
# X, y 값을 가지고 7:3 비율로 Train, Test 을 나누세요.
# train_test_split 시 stratify 옵션 사용

# 라이브러리 임포트
from sklearn.model_selection import train_test_split

# 데이터 Set 나누기 (Train:Test)
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.3, stratify=y, random_state=42)

print(X_train.shape, y_train.shape)
```

```python
# 5. 딥러닝 모델 구현 (DNN)
# ① 모델 생성하기 : Sequentioal 모델
# ② 모델 컴파일 하기
# ③ 모델 학습하기
# ④ Callbacks 함수 설정하기 : EarlyStopping, ModelCheckpoint

# ① 모델 생성하기 : Sequentioal 모델
# 하이퍼 파라미터 설정 : batch_size = 1024, epochs = 20
# hidden layer 3개 이상으로 모델을 구성
# 과적합 방지하는 dropout을 설정하세요.
# A. 이진 분류용 DNN layer
```

```python
# A. EarlyStopping
# 모델을 더 이상 학습을 못할 경우(loss, metric 등의 개선이 없을 경우), 학습 도중 미리 학습을 종료시키는 콜백 함수 입니다.

# monitor : EalryStopping의 기준이 되는 값('val_loss'를 입력하면 val_loss가 더 이상 감소되지 않을 경우 적용)
# min_delta : 개선된 것으로 간주하기 위한 최소한의 변화량
# patience : monitor되는 값의 개선이 없을 때 최적의 값을 기준으로 몇 번의 epoch를 더 진행할 지 정하는 값
# verbose : 화면에 상태 표시
# mode : monitor가 최소가 되어야 하는 지, 최대가 되어야 하는지 ('val_accuracy'는 max, 'val_loss'는 min)
```

```python
# [실습] 딥러닝 모델 구현하기
# [Q] 아래의 조건으로 DNN 모델을 구현해 보세요.
# 히든 레이어를 3개로 모델을 구성하고 과적합 방지를 위한 dropout 설정
# 모델의 optimizer는 'adam'을 사용하고 metrics는 'accuracy' 로 하여 컴파일하세요.
# epochos는 20으로 batch_size는 32로 설정
# EarlyStopping 콜백으로 정해진 epoch 동안 모니터링 지표가 향상되지 않을 때 훈련 중단( monitor는 'val_accuracy', patience는 5 설정)
# ModelCheckpoing 콜백으로 validation performance가 가장 좋은 모델을 '이름.h5' 파일로 저장하세요.
# 모델 학습 후 결과를 history_q 변수에 저장하세요.
```

```python
# 여기에 입력하세요
# 라이브러리 임포트
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import  EarlyStopping, ModelCheckpoint

# input layer 확인 
X_train.shape

# 모델 선언 (뱐수 : model_q)
model_q = Sequential()
model_q.add(Dense(32, activation='relu', input_shape=(80,)))
model_q.add(Dropout(0.3))
model_q.add(Dense(16, activation='relu'))
model_q.add(Dropout(0.3))
model_q.add(Dense(8, activation='relu'))
model_q.add(Dropout(0.3))
model_q.add(Dense(1, activation='sigmoid'))

# 모델 확인
model_q.summary()

# 모델 컴파일하기
model_q.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy']) 

# callbacks 함수 설정
es = EarlyStopping(monitor='val_accuracy', patience=5, mode='max', verbose=1)
mc = ModelCheckpoint('model_q/my_checkpoint_q.ckpt', monitor='val_accuracy', save_best_only=True, save_weights_only=True, verbose=1)

# 모델 훈련
history = model_q.fit(X_train, y_train, 
                        validation_data=(X_test, y_test),
                        epochs=20, 
                        callbacks=[es, mc],
                        batch_size=32,
                        verbose=1)

model_q.load_weights("model_q/my_checkpoint_q.ckpt")
model_q.save("석재민.h5")
```
